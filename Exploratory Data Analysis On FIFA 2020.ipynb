```python
#Exploratory Data Analysis on FIFA 2020 Players
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('players_20.csv')
```

```python
fifa.head()
fifa.shape
```

```python
for cols in fifa.columns:
    print(cols)
```

```python
values = fifa['nationality'].value_counts()[0:10]
```

```python
country = fifa['nationality'].value_counts()[0:10].keys()
```

```python
plt.figure(figsize = (10, 10))
plt.bar(country, values, color = 'g')
plt.ylabel('Values', size = 35)
plt.xlabel('Country', size = 35)
plt.show()
```

```python
player_salary = fifa[['short_name', 'wage_eur']]
```

```python
#player_salary.head()
player_salary = player_salary.sort_values(by = ['wage_eur'], ascending = False)
```

```python
player_salary.head()
```

```python
plt.figure(figsize = (8, 5))
plt.bar(list(player_salary['short_name'])[0:5], list(player_salary['wage_eur'])[0:5], color = 'black')
plt.show()
```
